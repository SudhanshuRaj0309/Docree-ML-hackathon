### Selecting top 100 cities

```python
term_counts = df['USERCITY'].value_counts()

# Select the top 100 most frequent terms
top_terms = term_counts.nlargest(100).index

# Set all other terms to 0
df.loc[~df['USERCITY'].isin(top_terms), 'USERCITY'] = "No CITY"
```

```python
URL_Keywords_dummies = pd.get_dummies(df['USERCITY'], prefix='USERCITY')
PLATFORM_ID_dummies = pd.get_dummies(df['PLATFORM_ID'], prefix='PLATFORM_ID')
df = pd.concat([df, PLATFORM_ID_dummies, URL_Keywords_dummies], axis=1)
```

```python
!pip install catboost

import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(tax_train, tax, test_size=0.2, random_state=42)

y_train = y_train.astype(str)
y_test = y_test.astype(str)

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

classifiers_catboost = []
i = 0
for class_label in label_encoder.classes_:
    print(i, '\n')
    i += 1
    # Create binary labels (1 for the class, 0 for others)
    y_train_binary = (y_train_encoded == label_encoder.transform([class_label])[0]).astype(int)

    classifier_catboost = CatBoostClassifier(
        iterations=5,
        learning_rate=0.4,
        depth=12,
        l2_leaf_reg=3,
        loss_function='Logloss',
        eval_metric='Accuracy',
        random_seed=0,
        task_type='GPU')
    classifier_catboost.fit(X_train, y_train_binary)
    classifiers_catboost.append(classifier_catboost)

y_test_probabilities_rf = []

y_test_probabilities_catboost = []
for classifier_catboost in classifiers_catboost:
    y_test_probabilities_catboost.append(classifier_catboost.predict_proba(X_test)[:, 1])

y_test_predicted_catboost = label_encoder.inverse_transform(np.argmax(np.column_stack(y_test_probabilities_catboost), axis=1))

accuracy_catboost = accuracy_score(y_test, y_test_predicted_catboost)
print("Accuracy (CatBoost):", accuracy_catboost)


import joblib
joblib.dump(classifiers_catboost, 'Cat_Taxonomy_model.pkl')
print("model saved!")
```
